```python
import pandas as pd

# Load dataset (adjust filename if needed)
df = pd.read_csv("/content/sleep_deprivation_dataset_detailed.csv")

# Preview data
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["Sleep_Hours"], df["Stroop_Task_Reaction_Time"])
plt.xlabel("Sleep Hours")
plt.ylabel("Stroop Reaction Time")
plt.title("Sleep Duration vs Cognitive Control (Stroop Task)")
plt.show()
```

```python
from scipy.stats import pearsonr

r, p = pearsonr(df["Sleep_Hours"], df["Stroop_Task_Reaction_Time"])

print("Correlation (r):", round(r, 3))
print("p-value:", round(p, 3))
```
